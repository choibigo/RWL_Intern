## 본격적으로 model 만들기!

```python
from torch import nn

class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fcs = nn.Sequential(nn.Linear(28*28, 100), # 입력과 첫 은닉층
                                 nn.ReLU(),
                                 nn.Linear(100, 10)) # 출력
    def forward(self, x):
        x = torch.flatten(x, start_dim=1) # flatten이 뭘까? 아래 보셈
        x = self.fcs(x)
        return x
```

```python
# flatten 확인
x_batch, _ = next(iter(train_DL))
print(x_batch.shape)

# 원래 [32, 1, 28, 28]을 [25088]으로 평탄화시킴
flat_images = torch.flatten(x_batch)
print(flat_images.shape)

# 이거는 첫 32를 빼고 그 뒤 부터 평탄화 <- 이걸 위에서 사용함.
flat_images2 = torch.flatten(x_batch, start_dim=1)
print(flat_images2.shape)
```

- 왜 이렇게 flatten을 하고 모델을 돌릴까?

  - 우리가 nn.Linear에서 입력층의 입력을 28*28으로 넣었다. 이는 28x28을 입력으로 받는게 아니라 당연히 784를 입력으로 받는거다.

- (중요) 모델과 데이터 모두 GPU에 올려야함.

```python
# MLP 모델 GPU 올리기
model=MLP().to(DEVICE)
print(model)

x_batch, _ = next(iter(train_DL)) # 데이터 한 국자 가져오기
print(x_batch.shape)

model.eval()
with torch.no_grad():
    # x_batch의 데이터들도 GPU에 올려줘야함. (테스트로 한번 집어넣어보는 것)
    print(model(x_batch.to(DEVICE)).shape)
    # 그래서 GPU VRAM이 작으면 배치도 작아야함!
```

## 모델 훈련시키기!

```python
from torch import optim

LR = 1e-3
EPOCH = 5
criterion = nn.CrossEntropyLoss() # CEE 손실함수 # 안에 Softmax를 포함하고 있음!! +++
optimizer = optim.Adam(model.parameters(), lr=LR) # 아담 옵티마이저

def Train(model, train_DL, criterion, optimizer, EPOCH):

    loss_history=[]
    NoT=len(train_DL.dataset) # The number of training data
    print(NoT)
    model.train() # train mode로 전환
    for ep in range(EPOCH):
        rloss = 0 # running loss
        for x_batch, y_batch in train_DL: # 데이터를 데이터로더 부터 꺼내기
            # 배치 데이터 GPU에 올리기 +
            x_batch = x_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)
            # (모델은 아까 올려서 지금은 안올려도 됨)

            # inference
            y_hat = model(x_batch)

            # loss
            loss = criterion(y_hat, y_batch)

            # update
            optimizer.zero_grad() # gradient 누적을 막기 위한 초기화
            loss.backward() # backpropagation
            optimizer.step() # weight update
            # loss accumulation
            loss_b = loss.item() * x_batch.shape[0] # batch loss # BATCH_SIZE를 곱하면 마지막 18개도 32개를 곱하니까..
            rloss += loss_b # running loss

        # print loss
        loss_e = rloss/NoT # epoch loss
        loss_history += [loss_e]
        print(f"Epoch: {ep+1}, train loss: {loss_e:.3f}")
        print("-"*20)

    return loss_history
```

- 아래쪽에 `loss_e = rloss/NoT `을 한 이유는 **제대로된 Loss 평균**을 구하고 싶어서임.
  - 예를 들어 총 데이터가 50개인데 배치가 32이면, 한쪽 배치는 32의 Loss 평균을 구하게되고, 나머지 한쪽은 18끼리의 평균임. 이게 서로 좀 다를 수도 있음.
  - 그래서 그냥 전체 Loss를 위 rloss에서 축적하고, 아래에서 Epoch (6만개의 데이터를 다 돌렸을 때) 총 데이터의 수 만큼 나누는 것이다.
  - 이를 실제로 쓰는 것은 아니고, 지표에 쓰이는 듯? 아래 **그래프 만드는데 쓰임**

- 실제 학습

```python
loss_history = Train(model, train_DL, criterion, optimizer, EPOCH)

plt.plot(range(1,EPOCH+1),loss_history)
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("Train Loss")
plt.grid()
```

## 모델 테스트하기

- 이 코드는 학습이 아니라 그냥 inference가 끝이라 조금 더 간단함.

```python
def Test(model, test_DL):
    model.eval() # test mode로 전환
    with torch.no_grad(): # grad 계산 ㄴㄴ
        rcorrect = 0
        for x_batch, y_batch in test_DL:
            x_batch = x_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)

            # inference
            y_hat = model(x_batch)

            # corrects accumulation
            pred = y_hat.argmax(dim=1) # 10 결과값 중에서 최고 뽑기
            # 진짜 잘 맞췄는지 대조하고 그 개수를 세기
            corrects_b = torch.sum(pred == y_batch).item() # torch.eq(pred, y_batch).sum().item()
            rcorrect += corrects_b # 이건 한 배치에 대해서 구한거니 계속 누적해야함.

        # 전체 test 데이터에 대한 정확도.
        accuracy_e = rcorrect/len(test_DL.dataset)*100
    print(f"Test accuracy: {rcorrect}/{len(test_DL.dataset)} ({accuracy_e:.1f} %)")
    return round(accuracy_e,1)
```

```python
Test(load_model, test_DL) # 우리가 가져온 load_model으로 테스트
```

## confusion matrix

- 딥러닝때 쓰기 좋은 그래프

```python
def get_conf(model, test_DL):
    N = len(test_DL.dataset.classes)
    model.eval()
    with torch.no_grad():
        confusion = torch.zeros(N,N)
        for x_batch, y_batch in test_DL:
            x_batch = x_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)
            # inference
            y_hat = model(x_batch)
            # accuracy
            pred = y_hat.argmax(dim=1)

            confusion += torch.bincount(N * y_batch.cpu() + pred.cpu(), minlength=N**2).reshape(N, N)
            # confusion matrix는 무조건 10x10이 되어야 하는데 만약 마지막 label에 대해 예측을 모두 실패하면 100개보다 작아질 수 있기 때문에 minlength를 설정

    confusion = confusion.numpy()

    return confusion
```

```python
confusion = get_conf(load_model, test_DL)
plt.imshow(confusion)
```

```python
import numpy as np

def plot_confusion_matrix(confusion, classes=None):
    N = confusion.shape[0]
    accuracy=np.trace(confusion)/np.sum(confusion) * 100

    # confusion = confusion/np.sum(confusion, axis=1)
    plt.figure(figsize=(10,7))
    plt.imshow(confusion, cmap="Blues")
    plt.title("confusion matrix")
    plt.colorbar()

    for i in range(N):
        for j in range(N):
            plt.text(j,i, round(confusion[i,j]),
                     horizontalalignment="center", fontsize=10,
                     color="white" if confusion[i,j] > np.max(confusion) / 1.5 else "black")

    if classes is not None:
        plt.xticks(range(N), classes)
        plt.yticks(range(N), classes)
    else:
        plt.xticks(range(N))
        plt.yticks(range(N))

    plt.xlabel(f"Predicted label \n accuracy = {accuracy:.1f} %")
    plt.ylabel("True label")
```

```python
confusion = get_conf(load_model, test_DL)
plot_confusion_matrix(confusion, test_DS.classes)
plt.xticks(rotation=45);
```

### 잠시 실험

MLP는 순서가 셔플된 데이터도 잘 학습할 수 있을까? 있다!

- 물론 셔플된 데이터를 처음부터 학습했다는 가정하에.

```python
# =========================================================
# [실험 1] 기존에 학습한 모델에 '섞인 이미지' 넣어보기
# =========================================================

# 1. 784(28x28)개의 픽셀 위치를 섞을 일관된 '규칙' 만들기
torch.manual_seed(42) # 동일한 섞임 규칙을 유지하기 위해 시드 고정
shuffle_idx = torch.randperm(28*28) # 0부터 783까지의 숫자를 무작위로 섞은 배열 생성

# 2. 이미지를 불러올 때마다 똑같은 규칙으로 섞어주는 변환기(Transform) 정의
class ShuffleTransform:
    def __call__(self, x):
        # x는 ToTensor()를 거친 후이므로 [1, 28, 28] 형태의 텐서입니다.
        x_flat = x.view(-1) # [784]로 한 줄로 쭉 펴기
        x_shuffled = x_flat[shuffle_idx] # 미리 만들어둔 규칙대로 픽셀 섞기
        return x_shuffled.view(1, 28, 28) # 다시 원래 형태인 [1, 28, 28]로 되돌리기

# 3. 데이터셋과 데이터로더 새로 정의하기 (섞기 기능 추가)
# 기존 transform(ToTensor) 뒤에 ShuffleTransform을 이어 붙입니다.
shuffled_transform = transforms.Compose([
    transforms.ToTensor(),
    ShuffleTransform()
])

# 실습하시던 코랩 드라이브 경로를 사용해 테스트 데이터셋 다시 불러오기
shuffled_test_DS = datasets.MNIST(root='/content/drive/MyDrive/Colab Notebooks/data',
                                  train=False, download=True, transform=shuffled_transform)
shuffled_test_DL = torch.utils.data.DataLoader(shuffled_test_DS, batch_size=BATCH_SIZE, shuffle=False)

# 4. 기존 모델(load_model)로 테스트 및 시각화해보기
print("--- 기존 모델에 섞인 이미지를 넣었을 때의 성능 ---")
Test(load_model, shuffled_test_DL)

# 기존에 만들어두신 Test_plot 함수 그대로 사용! (눈으로 섞인 이미지를 확인해보세요)
Test_plot(load_model, shuffled_test_DL)
```

```python
# =========================================================
# [실험 2] 섞인 이미지로 처음부터 '새로 학습'시켜보기
# =========================================================

# 1. 섞인 이미지로 구성된 '학습 데이터셋' 만들기
shuffled_train_DS = datasets.MNIST(root='/content/drive/MyDrive/Colab Notebooks/data',
                                   train=True, download=True, transform=shuffled_transform)
shuffled_train_DL = torch.utils.data.DataLoader(shuffled_train_DS, batch_size=BATCH_SIZE, shuffle=True)

# 2. 완전히 새롭게 뇌가 초기화된 MLP 모델 꺼내기
shuffled_model = MLP().to(DEVICE)

# 3. 새로운 옵티마이저 설정
shuffled_optimizer = optim.Adam(shuffled_model.parameters(), lr=LR)

# 4. 섞인 이미지로 학습 진행! (작성해두신 Train 함수 재사용)
print("\n--- 섞인 이미지로 새로운 모델 학습 시작 ---")
Train(shuffled_model, shuffled_train_DL, criterion, shuffled_optimizer, EPOCH)

# 5. 성능 평가 및 시각화
print("\n--- 섞인 이미지로 학습된 모델의 섞인 이미지 예측 성능 ---")
Test(shuffled_model, shuffled_test_DL)
Test_plot(shuffled_model, shuffled_test_DL)
```
