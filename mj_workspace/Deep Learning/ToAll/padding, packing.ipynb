### **Sequential data**

- 각 문장마다 구성된 단어의 개수가 다 다르다
- 따라서 두가지 방법 사용
    - **padding**
        - batch내에서 가장 긴 seq길이에 맞춰 나머지 짧은 seq들의 빈 공간을 <pad>또는 0으로 ㅊ움
            - 자원낭비 발생
        - Transformer사용

    - **Packing**
        - batch 내의 seq를 길이가 긴 순서대로 정렬
        - time-step별로 유효한 데이터만 묶어서 연산
            - 구현시 정렬 및 언패킹 과정 필요
