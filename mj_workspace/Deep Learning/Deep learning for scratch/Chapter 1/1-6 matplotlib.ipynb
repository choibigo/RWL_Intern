##### **1.6.1 단순한 그래프 그리기**

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)   # 0 ~ 5.9까지 0.1 간격으로 데이터 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x")  # 축이름
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()        # 범례 띄우기
plt.show()
```
