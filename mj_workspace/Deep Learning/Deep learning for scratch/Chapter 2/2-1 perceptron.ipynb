- 퍼셉트론(인공뉴런): 다수의 신호를 입력으로 받아 하나의 신호를 출력
- 신호: 흐른다(1), 흐르지 않는다(0) 두 가지 값을 가짐.
    - 입력신호: x1, x2
    - 출력신호: y
    - 가중치: w1, w2(w는 weight의 머리글자)

- 입력신호가 뉴런에 보내질 때 가각 고유한 가중치가 곱해짐. (w1x1, w2x2)

- 뉴런이 활성화: 뉴런에서 보내온 신호의 총합이 임계값을 넘어설 때만 1을 출력
    - 임계값: 정해진 한계. (세타 기호로 표현= b로 치환)
    - theta = bias로 치환하면 아래와 같이 표현가능


$$
y = \begin{cases} 
0 & (b + w_1x_1 + w_2x_2 \le 0) \\ 
1 & (b + w_1x_1 + w_2x_2 > 0) 
\end{cases}
$$

AND, NAND, OR은 아니까  Skip

##### **XOR**

- 한 쪽이 1일때만 1 출력
- 즉, **두 값이 다를 때만 1 출력**

---

##### **퍼셉트론 구현하기**


$$
y = \begin{cases} 
0 & (b + w_1x_1 + w_2x_2 \le 0) \\ 
1 & (b + w_1x_1 + w_2x_2 > 0) 
\end{cases}
$$

```python
import numpy as np
x = np.array([0, 1])     # input
w = np.array([0.5, 0.5]) # weight
b = -0.7                 # bias


np.sum(w*x)+b   # 2개의 input에 대한 출력 y 표현
```

OR을 weight와 bias로 표현시 
(b, w1, w2) = (-0.5, 1, 1)

- 회색 영역이 0을 출력하는 영역

- 즉, 1이 하나라도 포함되면 1을 출력하는 영역에 있다.

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

x1 = np.arange(-1, 3, 0.1)
x2 = -x1 + 0.5

plt.axvline(x=0, color = 'b')  # draw x =0 axes
plt.axhline(y=0, color = 'b')   # draw y =0 axes

# 그래프 그리기
plt.plot(x1, x2, label="or")
plt.xlabel("X1") # x축 이름
plt.ylabel("X2") # y축 이름
plt.legend()

plt.fill_between(x1, x2, -3, color='grey', alpha=None)

plt.scatter([0],[0],marker='o',color='r')
plt.scatter([1,0,1],[0,1,1],marker='^',color='r')
plt.show()
```

그러나 XOR의 경우 두 값이 다를 때만 1을 출력한다.

아래 사진을 보면 (-1,-1), (1,1)일 때 0을 출력하고

(1,-1), (-1,1)일때 1을 출력하는 영역을 **비선형 영역**으로 표현한 것이다.

이러한 XOR 게이트는 단층 퍼셉트론으로 표현이 불가하고 다층 퍼셉트론으로만 구현 가능하다.


```python
display(Image(filename='images/3.png', embed=True))
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

##### **정리**

- 퍼셉트론은 weight와 bias를 매개변수로 가짐.
- 단층 퍼셉트론은 선형 영역만 표현.
    - AND, OR, NAND 게이트 표현 가능
- 다층 퍼셉트론은 비선형 영역도 표현.
    - XOR 게이트 표현 가능
