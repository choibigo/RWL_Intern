- 퍼셉트론(인공뉴런): 다수의 신호를 입력으로 받아 하나의 신호를 출력
- 신호: 흐른다(1), 흐르지 않는다(0) 두 가지 값을 가짐.
    - 입력신호: x1, x2
    - 출력신호: y
    - 가중치: w1, w2(w는 weight의 머리글자)

- 입력신호가 뉴런에 보내질 때 가각 고유한 가중치가 곱해짐. (w1x1, w2x2)

- 뉴런이 활성화: 뉴런에서 보내온 신호의 총합이 임계값을 넘어설 때만 1을 출력
    - 임계값: 정해진 한계. (세타 기호로 표현= b로 치환)
    - theta = bias로 치환하면 아래와 같이 표현가능


$$
y = \begin{cases} 
0 & (b + w_1x_1 + w_2x_2 \le 0) \\ 
1 & (b + w_1x_1 + w_2x_2 > 0) 
\end{cases}
$$
