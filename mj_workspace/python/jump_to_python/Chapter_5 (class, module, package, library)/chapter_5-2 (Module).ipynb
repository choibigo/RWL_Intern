### **Module**

`함수, 변수, 클래스 등을 모아 둔 .py 파일` <br>
`(다른 파일에서 import 해서 사용 가능)`

* 코드가 길어질수록 한 파일에 다 넣으면 관리 불가
* 재사용 / 분리 / 테스트를 위해 기능 단위로 쪼갬
* 그 최소 단위가 모듈(module)

---
##### **기본구조**
```python
# mod1.py
def add(a, b):
    return a + b

def sub(a, b):
    return a - b
```
---

**1. 모듈 불러오기 (기본 import)** <br>

```python
import mod1

print(mod1.add(3, 4))
print(mod1.sub(4, 2))
```
<br>

**2. 모듈에서 함수 1개만 가져오기** <br>

```python
from mod1 import add

print(add(3, 4))
```
<br>

**3. 모듈에서 함수 여러개 가져오기** <br>

```python
from mod1 import add, sub
from mod1 import *         # `*`는 가급적 비권장 (모두 가져오는걸 뜻함)
```
