#### **print**




````python
print("life" "is" "too short")
print("life" + "is" + "too short")
# lifeistoo short

print("life", "is", "too short")  # 자동 공백 (=쉼표)
# life is too short

print("2025", "08", "17", sep="-")  # 공백을 `-`로 구분
# 2025-08-17


for i in range(5):
    print(i, end=" ")  # end 매개변수를 " "로 바꿈으로서 줄바꿈 제거
# 0 1 2 3 4
for i in range(5):
    print(i)       
# 0
# 1 
# ..

# f-string형식으로 출력 (권장)
a = 3
b = 4
print(f"{a} + {b} = {a + b}")
````

#### **파일 Read/Write**

* 결과를 영구 저장 (로그, 데이터, 설정)
* 사용자 입력보다 자동화·재현성이 높음

##### **파일 열기**

```py
파일객체 = open(파일이름, 모드)


| 모드  | 의미                |
| --- | ----------------- |
| `r` | 읽기 (파일 있어야 함)     |
| `w` | 쓰기 (파일 없으면 생성, 있으면 초기화) |
| `a` | 추가 (기존 내용 유지)     |


````python
f = open("test.txt", "w")
f.close()               # f.close()는 항상 호출 권장
````

##### **파일 쓰기**



````python
f = open("test.txt", "w")
for i in range(1, 4):
    f.write(f"{i}번째 줄\n")
f.close()
````

##### **파일 읽기**



````python

# 1. readlines() — 전체를 리스트로
f = open("test.txt", "r")
lines = f.readlines()
for line in lines:
    print(line.strip())
f.close()
````

````python
# 2. read() — 전체를 문자열로

f = open("test.txt", "r")
data = f.read()
print(data)
f.close()
````

````python
# 3. 파일 객체 + for 문
f = open("test.txt", "r")
for line in f:
    print(line.strip())
f.close()
````

---

#### **파일 내용 추가 (a)**

````python
f = open("test.txt", "a")
f.write("추가된 줄\n")
f.close()
````

---

#### **with문 (중요)**

* close() 자동 처리
* 예외 발생 시에도 안전

````python
with open("test.txt", "w") as f:
    f.write("Hello Python")

with open("test.txt", "r") as f:
    data = f.read()
    print(data)


````
