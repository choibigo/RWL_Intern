### **If문**

* 조건이 True 일때만 코드 실행
* 프로그램의 흐름 제어를 위해 사용


```python
if 조건:
    실행문
else:
    실행문
```
* `:` 필수
* 들여쓰기 필수(공백4칸 권장)


---

##### **조건문에 쓰이는 연산자 정리** 

**비교연산자**
`>, <, >=, <=, ==, !=`

**논리연산자**
`and, or, not`

**포함 여부**
`x in 리스트,
x not in 문자열`


---

if 조건1:

### **While문**

* 조건이 True인 동안 반복 실행


```python
while 조건문:
    실행문
```
* `:` 필수
* 들여쓰기 필수(공백4칸 권장)

- **반복 조건 + 반복을 멈추게 하는 변화(증가/감소)**가 함께 있어야 한다.
- 반복 변수의 값이 변하지 않으면 **무한 루프**에 빠지기 쉽다.



---

##### **Break**

* 반복문을 즉시 종료한다.
* 조건과 상관없이 가장 가까운 while 문을 빠져나간다.

```python
while True:
    if 조건:
        break   
```
---

##### **Break**

* 아래 코드를 실행하지 않고, 곧바로 **다음 반복(조건 검사)**으로 이동한다.

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```
---

##### **while True**

* 조건이 항상 참이므로 무한 루프 패턴이다.
* 반드시 break와 함께 사용해야 안전함

```python
while True:
    if 종료조건:
        break
```

---

##### **while - else**

* while 문이 break 없이 정상 종료되었을 때만 else 블록이 실행됨.
* break로 종료되면 else는 실행되지 않는다.

```python
count = 0
while count < 3:
    count += 1
else:
    print("정상 종료")
```
---

##### **중첩 while**

* while 안에 또 다른 while 사용 가능
* break / continue는 가장 가까운 while 문에만 영향을 준다.

```python
i = 1
while i <= 2:
    j = 1
    while j <= 3:
        print(i, j)
        j += 1
    i += 1
```
---
`while 문은 조건 + 변화 + 종료 제어(break/continue) 를 반드시 함께 설계하도록 권장.`

### **for문**

- **반복 가능한 객체(iterable)** 를 순서대로 하나씩 꺼내며 반복
- 리스트, 튜플, 문자열, range 등에서 사용

```python
for 변수 in 반복가능객체:
    실행문
```



---
#### **for vs while**

* for : 반복 횟수/대상이 명확할 때
* while : 반복 종료 조건이 동적으로 변할 때

````python
for x in [1, 2, 3]:
    print(x)
````

````python
for ch in "python":
    print(ch)
````

---
#### **range함수 (중요)**

* 숫자 반복에 가장 많이 사용 (마지막 값은 포함 X)


````python
range(10)        # 0 ~ 9
range(1, 11)     # 1 ~ 10
range(1, 10, 2)  # 1, 3, 5, 7, 9

for i in range(5):
    print(i)

````

---
#### **continue / break**

* continue : 아래 코드 건너뛰고 다음 반복
* break : 반복문 즉시 종료


````python
for i in range(10):
    print(i)
    if i == 5:
        break
````

````python
for i in range(5):
    if i % 2 == 0:
        continue
    print(i)
````

---
#### **for-else**

* break 없이 정상 종료되면 else 실행
* break로 종료되면 else 실행 X


````python
for i in range(5):
    print(i)
else:
    print("정상 종료")
````

---
#### **enumerate (인덱스 + 값)**

* 인덱스 관리할 때


````python
for idx, value in enumerate(['a', 'b', 'c']):   #반복할때 인덱스(번호)+값을 같이 주는 함수
    print(idx, value)
````

````python
for idx, value in enumerate(['a', 'b', 'c'], 1):  # 인덱스 번호를 1부터 시작하도록 함.
    print(idx, value)
````

---
#### **zip (여러 리스트 동시 순회)**

* 여러 리스트를 같은 인덱스 기준으로 묶음


````python
names = ['abc', 'def', 'qiowefjio']
scores = ['a', 'b', 'c']

for name, score in zip(names, scores):
    print(name, score)
````

#### **중첩 for**



````python
for i in range(2, 10):
    for j in range(1, 10):
        print(i * j)
````

````python
result = [x * 3 for x in [1, 2, 3, 4]]
print(result)
````
