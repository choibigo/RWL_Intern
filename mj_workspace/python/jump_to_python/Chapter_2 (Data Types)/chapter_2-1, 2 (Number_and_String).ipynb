### **문자열 인덱싱과 슬라이싱**

##### 인덱싱은 무언인가를 **가리킨다**라는 의미, 슬라이싱은 무엇인가를 **잘라낸다**라는 의미

Indexing : `문자열에서 특정 위치의 문자 하나를 선택. (0부터 카운트)`
```python
a[index_num]
```

````python
a = "Life is too short, You need Python"
print(a[0])    # 'L'
print(a[3])    # 'e'
print(a[-1])   # 'n'  (뒤에서 첫 번째 문자)
print(a[-0])   # 'L'  a[0]과 동일
````

Slicing : `문자열에서 범위를 잘라 새로운 문자열 생성. (끝 인덱스(end_num)는 포함하지 않음)` <br>
결과는 항상 새로운 문자열.

```python
a[start:end]
```
* start ≤ index < end
    * start 생략시 → 처음부터
    * end 생략시 → 끝까지

````python
a = "Life is too short, You need Python"
print(a[0:4])       # 0부터 3까지
print(a[0:])        # 0부터 끝까지
print(a[:17])       # 0부터 16까지
print(a[:])         # 전체 문자열

# 음수 인덱스를 활용한 슬라이싱
b = "0123456789"
print(b[3:-3])     # 3456 
````

````python
# 문자열은 Immutable(불변)하므로 슬라이싱 + 결합으로 수정해야한다.

# 틀린 방법
# a = "Pithon"
# a[1] = 'y'   # TypeError 발생
# print(a)

# 올바른 방법
a = "Pithon"
b = a[:1] + 'y' + a[2:]
print(b)
````

### **문자열 포매팅**

* **문자열 안에 변하는 값(숫자, 문자열, 변수 등)을 삽입**하는 기법
* 동일한 문장 구조에서 값만 바꿔 출력할 때 사용함

> ##### 1. (구식) % 포매팅


````python
"I ate %d apples for %s days." % (10, "three")       # %s는 모든 타입 자동 문자열 변환
# "Error is %d%%." % 98    # '%' 기호 출력을 위해서 %%

````

> ##### 2. str.format() 포매팅


````python
# 1. 여러 값 지정 (인덱스와 이름)
print("I ate {0} apples for {1} days.".format(10, "three"))
# 출력: I ate 10 apples for three days.

print("I ate {number} apples for {day} days.".format(number=10, day=3))
# 출력: I ate 10 apples for 3 days.


# 2. 정렬 (왼쪽, 오른쪽, 가운데)
print("{0:<10}".format("hi"))    # 왼쪽 정렬
# 출력: 'hi        '

print("{0:>10}".format("hi"))    # 오른쪽 정렬
# 출력: '        hi'

print("{0:^10}".format("hi"))    # 가운데 정렬
# 출력: '    hi    '


# 3. 공백 채우기
print("{0:=^10}".format("hi"))   # 가운데 정렬 + '=' 채우기
# 출력: '====hi===='

print("{0:!<10}".format("hi"))   # 왼쪽 정렬 + '!' 채우기
# 출력: 'hi!!!!!!!!'


# 4. 소수점 표현
print("{0:.4f}".format(3.42134234))    # 소수점 4자리
# 출력: '3.4213'

print("{0:10.4f}".format(3.42134234))  # 10칸 확보 후 소수점 4자리
# 출력: '    3.4213'


# 5. 중괄호 { } 문자 자체 출력
print("{{ and }}".format())
# 출력: '{ and }'
````

> #### 3. f-string (권장 방식, Python 3.6+)
`f"~~{변수} "`

* .format()보다 속도가 빠르고, 가독성이 좋다.
* `f"{name.upper()}"`처럼 중괄호 안에서 파이썬 함수나 연산도 바로 쓸 수 있다.

````python
name = "홍길동"
age = 30
f"이름: {name}, 나이: {age}, 나이+1: {age+1}" 
````

````python
# 딕셔너리
d = {"name": "홍길동", "age": 30}
f"{d['name']}은 {d['age']}살"
````

````python
# 1. 정렬 (Alignment)
print(f"{'hi':<10}")      # 왼쪽 정렬 (10칸)
# 출력: 'hi        '

print(f"{'hi':>10}")      # 오른쪽 정렬 (10칸)
# 출력: '        hi'

print(f"{'hi':^10}")      # 가운데 정렬 (10칸)
# 출력: '    hi    '


# 2. 채우기 (Filling)
print(f"{'hi':=^10}")     # 가운데 정렬 + 빈 공간 '=' 채우기
# 출력: '====hi===='

print(f"{'hi':!<10}")     # 왼쪽 정렬 + 빈 공간 '!' 채우기
# 출력: 'hi!!!!!!!!'


# 3. 소수점 및 숫자 포맷 (Number Formatting)
print(f"{3.42134234:.4f}")    # 소수점 4자리까지 표시
# 출력: '3.4213'

print(f"{3.42134234:10.4f}")  # 10칸 확보하고 소수점 4자리 표시
# 출력: '    3.4213'

print(f"{1500000:,}")         # 천 단위 콤마(,) 넣기
# 출력: '1,500,000'


# 4. 중괄호 { } 문자 출력
print(f"{{ and }}")
# 출력: '{ and }'
````

### **문자열 관련 함수들**

문자열 자료형은 자체적으로 다양한 내장 함수를 제공하고, <br>
모든 문자열 함수는 *원본 문자열을 변경하지 않고 새로운 문자열(또는 값)을 반환*한다.

````python
# 기준 문자열
text = "Life is too short"
space_text = "     Strip test"
data = "a:b:c:d"
word = "hobby"
num1 = "12345"
num2 = "12 34"
alpha = "Python"
mix = "Python3"

# 1. 개수 세기 (count)
print(word.count('b'))          # 2   [b의 개수를 count]

# 2. 위치 찾기 (find, index)
print(text.find('too'))          # 8   [too의 처음 등장하는 인덱스 위치]
print(text.find('hello'))        # -1   [못찾으면 -1]

print(text.index('Life'))        # 0
# print(text.index('hello'))     # ValueError (없으면 오류)

# 3. 공백 제거 (strip)
stripped = space_text.strip()         # [새로운 변수에 저장 필수]
print(stripped)                  # "Strip test" 

# 4. 대소문자 변환 (upper, lower)
print(stripped.upper())          # "STRIP TEST"
print(stripped.lower())          # "strip test"

# 5. 문자열 치환 (replace)
replaced = stripped.replace("Strip", "for")
print(replaced)                  # "for test"

# 6. 문자열 분리 (split)
words = stripped.split()         # 모든 공백문자 (space, tab, newline 기준으로 나눔)
print(words)                     # ['Strip', 'test']

items = data.split(':')          # `:`을 기준으로 구분
print(items)                     # ['a', 'b', 'c', 'd']

# 7. 문자열 결합
joined = ",".join(items)         # 각 요소사이 `,` 삽입하여 결합
print(joined)                    # "a,b,c,d"

# 8. 문자열 검사 (알파벳으로만 구성되어있는지 검사)
print(alpha.isalpha())           # True
print(mix.isalpha())             # False

# (문자열로만 이루어진지 검사)
print(num1.isdigit())            # True
print(num2.isdigit())            # False

# 9. 시작 / 끝 검사
print(stripped.startswith("Life"))   # False
print(stripped.endswith("test"))    # True

# 10. immutable 확인
original = "hi"
original.upper()
print(original)                  # "hi"  (오류)

# 함수를 통해 생성한 문자열을 사용하기 위해서는 새로운 변수에 재할당 해주어야 함 (필수)
original = original.upper()
print(original)                  # "HI"  (good)
````
