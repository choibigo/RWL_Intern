### **Bool 자료형**

* 참(True) / 거짓(False) 을 나타내는 자료형
* 값은 오직 2개
    * True
    * False

> 반드시 대문자 사용 (true, false 불가)



````python
a = True
b = False
type(a)   # bool
type(b)   # bool
````

````python
# ---------- 비교 연산 결과 : Bool 로 표현----------
1 == 1    # True
2 > 1     # True
2 < 1     # False
````

````python
# ---------- 자료형의 참(True) / 거짓(False)----------
# True
"python"
[1, 2, 3]
(1, 2)
{'a': 1}
1, -1, 3.14

# False
""      # 빈 문자열
[]      # 빈 리스트
()      # 빈 튜플
{}      # 빈 딕셔너리
0       # 숫자 0
None

# 즉, 비어 있으면 False, 값이 있으면 True
````

````python
# ---------- 조건문에서의 활용----------
if []:
    print("참")
else:
    print("거짓")
# 거짓
````

````python
# ---------- while문에서의 활용----------
a = [1, 2, 3, 4]

while a:
    print(a.pop())   # 리스트에 값이 있는 동안 : True
                     # 비어지면 False : 반복종료
````

````python
# ---------- bool() 함수----------
# 값을 명시적으로 True/False로 반환
bool("python")   # True
bool("")         # False
bool([1,2,3])    # True
bool([])         # False
bool(0)          # False
bool(3)          # True=
````

````python
# ---------- 논리연산자----------
# AND
print("True and True  :", True and True)     # True
print("True and False :", True and False)    # False
print()

# OR
print("True or False  :", True or False)     # True
print("False or False :", False or False)    # False
print()

# NOT
print("not True :", not True)    # False
print("not False:", not False)   # True
print("not 1    :", not 1)       # False
print("not 0    :", not 0)       # True
print()
````

````python
# 8. 실전 예제

x = 5
y = 10

print("x > 0 and y > 0 :", x > 0 and y > 0)    # True
print("x > 10 or y > 5 :", x > 10 or y > 5)    # True
print("not (x > y)     :", not (x > y))        # True
````
