### **Tuple (튜플) 자료형**

튜플(tuple)은 리스트와 매우 유사하지만, <br>
한 번 생성하면 요소를 변경할 수 없는(immutable) 자료형

List는 [], Tuple은 ()


| 구분       | 리스트(List)    | 튜플(Tuple)         |
| -------- | ------------ | ----------------- |
| 기호       | `[]`         | `()`              |
| 수정       | 가능 (mutable) | ❌ 불가능 (immutable) |
| 요소 추가/삭제 | 가능           | ❌ 불가능             |
| 사용 목적    | 값 변경이 잦은 경우  | 값 고정이 필요한 경우      |


````python
# ---------- Tuple 생성 방법----------
t1 = ()
t2 = (1,)
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ('a', 'b', ('ab', 'cd'))

# ---------- 주의사항! ----------
# 요소가 1개일때 `,`를 붙여주어야 tuple.
t = (1)     # int
t = (1,)    # tuple
# 소괄호는 생략가능
t = 1, 2, 3   # tuple

# ---------- Tuple은 Immuntable ----------
# 요소 삭제 불가
t = (1, 2, 'a', 'b')
del t[0]      # TypeError

# 요소 변경 불가
t[0] = 'c'    # TypeError
````

---

#### **Tuple 연산**

````python
# ------- 인덱싱 ---------
t = (1, 2, 'a', 'b')
print(t[0])   # 1
print(t[3])   # 'b'

# ------- 슬라이싱 ---------
t = (1, 2, 'a', 'b')
print(t[1:])     # (2, 'a', 'b')

# ------- 더하기 ---------
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
t3 = t1 + t2          # immutable하므로 새 튜플에 할당해야함
print(t3)
# (1, 2, 'a', 'b', 3, 4)

# ------- 곱하기 ---------
t = (3, 4)
print(t * 3)
# (3, 4, 3, 4, 3, 4)

# ------- 길이(len) ---------
t = (1, 2, 'a', 'b')
len(t)   # 4
````

#### **Tuple에 없는 기능**

Tuple은 immutable이므로 아래 메서드 없음.

* append
* insert
* remove
* pop
* sort
* reverse
