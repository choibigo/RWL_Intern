### **Variables**

* 변수는 값을 저장하지 않는다
* 객체(Object)를 가리킨다(참조)


##### **변수명 규칙**
* 영문자, 숫자, 언더스코어(_)만 사용할 수 있다.
* 숫자로 시작할 수 없다.
* 예약어는 사용할 수 없다.
* 대소문자를 구분한다.


---

##### **참조 vs 복사 (중요)** 

````python
# a is b → True
a = [1, 2, 3]
b = a        # 같은 객체 (하나 바꾸면 둘다 바뀜!!)

# a is b → False
b = a.copy()       # 또는 a[:]
````

---

##### **여러 변수 대입** 

````python
a, b = 1, 2
a = b = 10
a, b = b, a   # 값 교환
````
