### **List 자료형**

* 리스트는 여러 값을 순서대로 저장하는 자료형이다.
* 숫자, 문자열, 다른 리스트 등 모든 자료형을 요소로 가질 수 있으며,
* 문자열과 달리 값 수정이 가능(mutable) 하다.


````python
# 빈 리스트 생성
a = []
a = list()

# 리스트 생성 
b = [1, 2, 3]
d = [1, 2, 'Life', 'is']
e = [1, 2, ['Life', 'is']] # 이중 리스트 형태 

# List Indexing
print("-------Indexing-------")
print(a)        # []
print(b)        # [1,2,3]
print(d[3])     # is
print(e[1])     # 2
print(e[2])     # ['Life', 'is']
print(e[2][0])  # Life                  -> {중첩 리스트 인덱싱}

# List Slicing
print("-------Slicing-------")
print(b[0:])    # [1,2,3]
print(b[:2])    # [1,2]
print(e[1:4])   # [2, ['Life', 'is']]  -> {중첩 리스트 Slicing}
print(e[2][:1]) # ['Life']             -> {중첩 리스트 Slicing}
````

##### **List 연산**

````python
# 더하기 
print([1, 2, 3] + [4, 5, 6])
# [1, 2, 3, 4, 5, 6]

# 반복
print([1, 2, 3] * 3)
# [1, 2, 3, 1, 2, 3, 1, 2, 3]

# 길이 (len)
len([1, 2, 3])   # 3
````

---

##### **List 수정과 삭제 (Mutable)**

````python
# ------------값 수정------------
a = [1, 2, 3]
a[2] = 4
print (a)       # [1, 2, 4]

# ------------삭제 (del)------------
del a[1]
print(a)        # [1, 4]

del a[1:]       # 인덱스 번호(1)부터 뒤까지 모두 del
print(a)
````

##### **List 관련 함수**

````python
# 기준 리스트
a = [1, 2, 3, 1]
print("초기 리스트:", a)
# 초기 리스트: [1, 2, 3, 1]

# ---------1. append : 요소 추가 (맨 뒤)---------
a.append(4)
print("append(4):", a)
# append(4): [1, 2, 3, 1, 4]

# 리스트도 요소로 추가 가능
a.append([5, 6])
print("append([5, 6]):", a)
# append([5, 6]): [1, 2, 3, 1, 4, [5, 6]]

# --------- 2. insert : 원하는 위치에 삽입(첫 번째 기준)---------
a.insert(1, 100) 
print("insert(1, 100):", a)
# insert(1, 100): [1, 100, 2, 3, 1, 4, [5, 6]]

# --------- 3. remove : 값으로 제거 (첫 번째만) ---------
a.remove(1)
print("remove(1):", a)
# remove(1): [100, 2, 3, 1, 4, [5, 6]]

# --------- 4. pop : 꺼내면서 제거 ---------
popped = a.pop()         # 마지막 인덱스 리스트 요소 반환 및 제거
print("pop():", popped, "| 결과:", a)
# pop(): [5, 6] | 결과: [100, 2, 3, 1, 4]

popped_index = a.pop(1)   # 인덱스(1) 값 반환 및 제거
print("pop(1):", popped_index, "| 결과:", a)
# pop(1): 2 | 결과: [100, 3, 1, 4]

# --------- 5. index : 값의 위치 찾기 ---------
print("index(3):", a.index(3)) 
# index(3): 1

# --------- 6. count : 값 개수 세기 ---------
print("count(1):", a.count(1))
# count(1): 1

# --------- 7. sort : 정렬 (원본 변경) --------- 
a.sort()
print("sort():", a)
# sort(): [1, 3, 4, 100]

# --------- 8. reverse : 순서 뒤집기 ---------
a.reverse()
print("reverse():", a)
# reverse(): [100, 4, 3, 1]

# --------- 9. extend : 리스트 확장 ---------
a.extend([7, 8])
print("extend([7, 8]):", a)
# extend([7, 8]): [100, 4, 3, 1, 7, 8]

# --------- 10. del : 인덱스로 삭제 --------- 
del a[1]
print("del a[1]:", a)
# del a[1]: [100, 3, 1, 7, 8]

# --------- 슬라이싱 삭제 ---------
del a[2:]
print("del a[2:]:", a)
# del a[2:]: [100, 3]

````
