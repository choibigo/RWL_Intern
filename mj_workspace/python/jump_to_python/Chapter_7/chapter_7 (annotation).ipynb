### **파이썬 타입 어노테이션**

동적 언어의 유연함은 유지하되, 코드 실행 전(Static)에 타입 오류를 미리 잡아 안정성과 가독성을 높이는 안전장치이다.
 
| 구분    | 동적(파이썬)     | 정적(자바 등)   |
| ----- | ----------- | ---------- |
| 타입 결정 | 런타임         | 컴파일 타임     |
| 타입 변경 | 가능          | 불가         |
| 장점    | 빠른 개발/유연    | 안전/대규모에 강함 |
| 단점    | 규모 커질수록 버그↑ | 개발 속도/유연성↓ |


##### **변수 어노테이션 방법**

```python
# 변수 타입 어노테이션
num: int = 1
name: str = "홍길동"
numbers: list[int] = [1, 2, 3]

# 함수 어노테이션
def add(a: int, b: int) -> int:
    return a + b
```

#### **Typing 모듈로 하기 (뭔가 코드에서 많이 봤었던 형태임)**

```python
# Optional (None 가능)
from typing import Optional
user_name: Optional[str] = None   # str | None

# Union (여러 타입 가능)
from typing import Union
user_id: Union[int, str] = "jenny"

# Callable (함수 타입)
from typing import Callable
def process(cb: Callable[[int], str]) -> str:
    return cb(42)

# Any (타입 검사 포기)
from typing import Any
data: Any = {"key": "value"}



---

#### **결론**

* 파이썬은 끝까지 동적 언어다

* 타입 어노테이션을 써도 런타임에서 타입을 강제하지 않는다.

* 코드 실행 결과에는 영향이 없다.
