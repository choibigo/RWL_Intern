## 2.3 퍼셉트론 구현하기

### 2.3.1 간단한 구현부터

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

AND 게이트 테스트

```python
AND(0, 0)
```

```python
AND(1, 0)
```

```python
AND(0, 1)
```

```python
AND(1, 1)
```

### 2.3.3 가중치와 편향 구현하기

***and_gate.py***

```python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

***nand_gate.py***

```python
import numpy as np

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])  # AND와는 가중치(w와 b)만 다르다.
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

***or_gate.py***

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])  # AND와는 가중치(w와 b)만 다르다!
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

## 2.5 다층 퍼셉트론이 출동한다면

### 2.5.2 XOR 게이트 구현하기

***xor_gate.py***

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2) # 둘다 T 인경우 F
    s2 = OR(x1, x2) # 둘중 하나만 T 인경우 T
    y = AND(s1, s2) # 둘다 T 인경우 T
    return y
```

```python
XOR(0, 0)
```

```python
XOR(1, 0)
```

```python
XOR(0, 1)
```

```python
XOR(1, 1)
```
