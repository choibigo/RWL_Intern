Hypothesis

$$
H(X) = \frac{1}{1 + e^{-W^T X}}
$$


- Binary Classification 문제를 풀기 위해 사용
- 데이터 $X$를 넣었을 때, 결과가 1일 확률을 계산하는 함수

Cost

$$
\text{cost}(W)
= -\frac{1}{m} \sum \Big( y \log(H(x)) + (1 - y)\log(1 - H(x)) \Big)
$$


- If $y \approx H(x)$, cost is near 0.
- If $y \ne H(x)$, cost is high.


Weight Update via Gradient Descent

$$
W := W - \alpha \frac{\partial}{\partial W} \text{cost}(W)
$$


- $\alpha$: Learning rate
