datetime.date

```python
import datetime
day1 = datetime.date(2021, 12, 14)
day2 = datetime.date(2023, 4, 5)
print(day1)
print(type(day1))

diff = day2 - day1
print(diff.days)
```

time.time

```python
import time
a = time.time()
for i in range(1,1000):
    print("")
b = time.time()
print(b-a)
```

time.sleep

```python
import time
for i in range(10):
    print(i)
    time.sleep(1)
```
