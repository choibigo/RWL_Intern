04-3 파일 읽고 쓰기

파일 생성하기

```python
f = open("doit/새파일.txt", 'w')
f.close()
```

```python
f = open("C:/doit/새파일.txt", 'w')
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

파일을 쓰기 모드로 열어 내용 쓰기

```python
f = open("새파일.txt", 'w', encoding = "UTF-8")
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

파일을 읽는 여러 가지 방법

readline 함수 이용하기

```python
f = open("C:/doit/새파일.txt", 'r')
line = f.readline()
print(line)
f.close()
```

```python
f = open("C:/doit/새파일.txt", 'r')
while True:
    line = f.readline()
    if not line: # 문자열은 참, 문자열이 없으면 거짓
        break 
    print(line)
f.close()
```

```python
while True:
    data = input()
    # data = int(data)
    print(type(data))
    if not data: 
        break
    print(data)
```

readlines 함수 사용하기

```python
f = open("C:/doit/새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

줄 바꿈(\n) 문자 제거하기

```python
f = open("C:/doit/새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
    print(line)
f.close()
```

```python
f = open("C:/doit/새파일.txt", 'r')
for line in f:
    print(line)
f.close()
```

파일에 새로운 내용 추가하기

```python
f = open("C:/doit/새파일.txt",'a') # w 모드로 사용하게 되면 그냥 덮어써버림
for i in range(11, 20):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

with 문과 함께 사용하기

```python
f = open("foo.txt", 'w')
f.write("Life is too short, you need python")
f.close()
```

```python
with open("foo.txt", "w") as f:
    f.write("Life is too short, you need python")
    # with 구문 안쪽의 f 는 지역변수이다.
```
