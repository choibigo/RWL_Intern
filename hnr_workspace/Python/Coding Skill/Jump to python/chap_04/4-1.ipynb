입력값이 몇 개가 될지 모를 때는 어떻게 해야 할까?

여러 개의 입력값을 받는 함수 만들기

```python
def add_many(*args): 
    result = 0 
    for i in args: 
        result = result + i   # *args에 입력받은 모든 값을 더한다.
    return result 

print(add_many(1,2,3,4,5))
```

```python
def add_mul(choice, *args): 
     if choice == "add":   # 매개변수 choice에 "add"를 입력받았을 때
         result = 0 
         for i in args: 
             result = result + i 
     elif choice == "mul":   # 매개변수 choice에 "mul"을 입력받았을 때
         result = 1 
         for i in args: 
             result = result * i 
     return result 

result1 = add_mul("add", 1,2,3)
result2 = add_mul("mul", 1,2,3)
print(result1)
print(result2)
```

함수의 반환값은 언제나 하나이다

```python
def add_and_mul(a,b): 
    return a+b, a*b

print(add_and_mul(3,4))
# 하나의 튜플을 리턴한다.
```

- add_and_mul(2, 3)의 반환값은 5 하나뿐이다. 두 번째 return 문인 return a * b는 실행되지 않았다는 뜻이다. 즉, 함수는 return 문을 만나는 순간, 반환값을 반환한 다음 함수를 빠져나가게 된다.

```python
def add_and_mul(a,b): 
    return a+b
    return a*b # return 이 한번 나오면 함수 자체가 종료되어 버리기 때문에 2개가 존재할 수 없다.

print(add_and_mul(3,4))
```

return 의 또 다른 쓰임새

- 특별한 상황일 때 함수를 빠져나가고 싶다면 return을 단독으로 써서 함수를 즉시 빠져나갈 수 있다.

```python
def say_nick(nick): 
    if nick == "바보": 
        return 
    print("나의 별명은 %s 입니다." % nick)

say_nick("김한누리")
say_nick("조코딩님")
say_nick("바보")
```

매개변수에 초깃값 미리 설정하기

```python
def say_myself(name, age, man=True): 
    print("나의 이름은 %s 입니다." % name) 
    print("나이는 %d살입니다." % age) 
    if man: 
        print("남자입니다.")
    else: 
        print("여자입니다.")

say_myself("김한누리", 28, False) # 기본값으로 True 가 들어가게 됨
```

함수 안에서 선언한 변수 효력 범위

```python
a = 1 # 전역 변수 a

def vartest(a):
    a = a +1

# 지역 변수 a -> 함수 안쪽에서 선언된 a : 함수가 끝나면 지역변수는 더 이상 쓰이지 않는다.
vartest(a)
print(a)
```

```python
def vartest(a):
    a = a + 1

vartest(3)
print(a)
```

함수 안에서 함수 밖의 변수를 변경하는 방법

1. return 사용하기

```python
a = 1 
def vartest(x): 
    x = x +1 
    return x

vartest(a)
```

2. Global 명령어 사용하기

```python
a = 1 
def vartest(): 
    global a # 전역변수
    a = a+1

vartest() 
print(a)
```
