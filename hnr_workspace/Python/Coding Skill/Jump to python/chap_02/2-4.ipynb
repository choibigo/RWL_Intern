튜플 요솟값을 삭제하려 할 때

```python
t1 = (1, 2, 'a', 'b')
del t1[0]
```

튜플 요소값을 변경하려 할 때

```python
t1 = (1, 2, 'a', 'b')
t1[0] = 'c'
```

인덱싱하기

```python
t1 = (1, 2, 'a', 'b')
print(t1[0])
print(t1[3])
```

슬라이싱하기

```python
t1 = (1, 2, 'a', 'b')
print(t1[1:])
```

튜플 더하기

```python
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
t3 = t1 + t2
print(t3)
```

튜플 곱하기

```python
t2 = (3, 4)
t3 = t2 * 3
print(t3)
```
