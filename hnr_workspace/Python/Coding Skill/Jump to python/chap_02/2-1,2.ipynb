f 문자열 포매팅

```python
name = '홍길동'
age = 30
a = f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
print(a)
```

```python
age = 30
a = f'나는 내년이면 {age + 1}살이 된다.'
print(a)
```

```python
y = 3.42134234
a = f'{y:0.4f}'  # 소수점 4자리까지만 표현
b = f'{y:10.4f}'  # 소수점 4자리까지 표현하고 총 자리수를 10으로 맞춤
print(a)
print(b)
```
