리스트의 슬라이싱

```python
a = [1, 2, 3, 4, 5]
print(a[0:2:1])
print(a[0:2:2])
print(a[0::2])
```

```python
a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]
print(a[2:5])
print(a[3][:2])
```

리스트 반복하기

```python
a = [1, 2, 3]
print(a * 3)
```

```python
a = [1, 2, 3]
print(a[2] + "hi") # 3이라는 숫자와 문자열을 더한다? 서로 다른 자료형은 더해지지 않음
```

리스트의 수정과 삭제

리스트의 값 수정하기

```python
a = [1, 2, 3]
a[2] = 4
a
```

del 함수를 사용해 리스트 요소 삭제하기

```python
a = [1, 2, 3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

리스트 관련 함수

리스트에 요소 추가하기 - append

```python
a = [1, 2, 3]
a.append(4)
a.append([5,6]) # 4번 index 에 list 가 들어간 형태
a
```

리스트 정렬 - sort

```python
a = [1, 4, 3, 2]
a.sort()
print(a)
```

리스트 뒤집기 - reverse

```python
a = ['a', 'c', 'b']
a.reverse()
print(a)
```

```python
a = ['a', 'c', 'b']
a.sort()
a.reverse()
print(a)
```

인덱스 반환 - index

```python
a = [1, 2, 3]
print(a.index(3)) #3 이 몇번째 인덱스 인가?
```

리스트 요소 삽입 - insert

```python
a = [1, 2, 3]
a.insert(1, 4) # 몇번째 인덱스에 어떠한 값을 넣겠다.
a
```

리스트 요소 제거 - remove

```python
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
a.remove(3)
a
```

리스트 요소 끄집어 내기 - pop

```python
a = [1, 2, 3]
print(a.append(4))
# 맨 끝을 날려버린다
print(a.pop())
print(a)
```

```python
a = [1, 2, 3]
print(a.pop(1))
print(a)
```

리스트에 포함된 요소 x의 개수 세기 - count

```python
a = [1, 2, 3, 1]
a.count(1)
```

리스트 확장 - extend

```python
a = [1, 2, 3]
a.extend([4, 5])
print(a)
```

```python
a = [1, 2, 3]
a.append([4, 5])
print(a)
```

```python
a = [1, 2, 3]
a.extend([4, 5])
print(a)
b = [6, 7]
a.extend(b)
print(a)
```
