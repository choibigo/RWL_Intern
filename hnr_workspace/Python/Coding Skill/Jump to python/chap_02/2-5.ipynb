딕셔너리 쌍 추가하기

```python
a = {1: 'a'}
a[2] = 'b'
print(a)
```

```python
a['name'] = 'pey'
print(a)
```

```python
a[3] = [1, 2, 3]
a
```

딕셔너리 요소 삭제하기

딕셔너리에서 Key 를 사용해 Value 얻기

```python
grade = {'pey': 10, 'julliet': 99}
print(grade['pey'])
print(grade['julliet'])
```

```python
a = {1:'a', 2:'b'}
print(a[1])
print(a[2])
```

```python
a = {'a': 1, 'b': 2}
print(a['a'])
print(a['b'])
```

```python
dic = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print(dic['name'])
print(dic['phone'])
print(dic['birth'])
```

딕셔너리 관련 함수

key 리스트 만들기 - keys

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(a.keys())
```

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(list(a.keys()))
```

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
for k in a.keys():
    print(k)
```

```python
print(list(a.keys()))
```

Value 리스트 만들기 - values

```python
a.values()
```

Key, Value 쌍 얻기 -items

```python
a.items()
```

Key : Value 쌍 모두 지우기 - clear

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
a.clear()
a
```

Key로 Value 얻기 - get

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(a.get('name'))
print(a.get('phone'))
```

```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print(a.get('hi'))
print(a['hi'])
```

```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print(a.get('hi', " 값이 없습니다."))
```

해당 Key가 딕셔너리 안에 있는지 조사하기 - in


```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print('name' in a)
print('email' in a)
```

Key로 Value 얻기 - pop

```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
phone = a.pop('phone')
print(phone)
print(a)
```
