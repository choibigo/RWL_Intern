딕셔너리 쌍 추가하기

```python
a = {1: 'a'}
a[2] = 'b'
print(a)
```

```python
a['name'] = 'pey'
print(a)
```

```python
a[3] = [1, 2, 3]
a
```

딕셔너리 요소 삭제하기

```python
del a[1]
print(a)
```

딕셔너리에서 Key 를 사용해 Value 얻기

```python
grade = {'pey': 10, 'julliet': 99}
print(grade['pey'])
print(grade['julliet'])
```

```python
a = {1:'a', 2:'b'}
print(a[1])
print(a[2])
```

```python
a = {'a': 1, 'b': 2}
print(a['a'])
print(a['b'])
```

```python
dic = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print(dic['name'])
print(dic['phone'])
print(dic['birth'])
```

딕셔너리를 만들 때 주의할 사항

```python
a = {1:'a', 1:'b'}
print(a)
```
