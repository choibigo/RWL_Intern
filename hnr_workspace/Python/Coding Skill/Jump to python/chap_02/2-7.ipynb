자료형의 참과 거짓

```python
a = [1, 2, 3, 4]
while a:
    print(a)
    a.pop()
```

```python
if []:
     print("참")
else:
     print("거짓")
```

```python
if [1,2,3]:
    print("참")
else:
    print("거짓")
```
