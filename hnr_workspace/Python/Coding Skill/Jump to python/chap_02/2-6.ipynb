교집합, 합집합, 차집합 구하기

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
```

```python
s1 & s2
```

```python
s1.intersection(s2)
```

```python
s1 | s2
```

```python
s1.union(s2)
```

```python
print(s1 - s2)
print(s2 - s1)
```

```python
print(s1.difference(s2))
print(s2.difference(s1))
```

집합 자료형 관련 함수

값 1개 추가하기 - add

```python
s1 = set([1, 2, 3])
s1.add(4)
print(s1)
```

값 여러개 추가하기 - update

```python
s1 = set([1, 2, 3])
s1.update([4, 5, 6])
print(s1)
```

특정 값 제거하기 - remove

```python
s1 = set([1, 2, 3])
s1.remove(2)
print(s1)
```

특정 값 제거하기 - discard

```python
s1 = set([1, 2, 3])
s1.discard(2)
print(s1)
s1.discard(4)
print(s1)
```

모든 값 제거하기 - clear

```python
s1 = set([1, 2, 3])
s1.clear()
print(s1)
```
