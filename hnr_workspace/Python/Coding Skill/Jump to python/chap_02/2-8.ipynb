리스트를 복사하고자 할 때

```python
a = [1, 2, 3]
b = a
```

```python
print(id(a))
print(id(b))
print(a is b) # 주소 값이 같은지 본다
```

```python
a = [1, 2, 3]
b = a
a[1] = 4
print(a)
print(b)
```

1. [:] 이용하기

```python
a = [1, 2, 3]
b = a[:]
a[1] = 4
print(a)
print(id(a))
print(b)
print(id(b))
```

2. copy 모듈 이용하기

```python
from copy import copy
a = [1, 2, 3]
b = copy(a)
```

```python
b is a 
```

```python
a = [1, 2, 3]
b = a.copy()
b is a
```

변수를 만드는 여러 가지 방법

```python
a, b = ('python', 'life')
print(a)
print(b)
```

```python
a, b = 'python', 'life'
print(a)
print(b)
```

```python
(a, b) = 'python', 'life'
print(a)
print(b)
```

```python
(a, b) = ('python', 'life')
print(a)
print(b)
```

```python
[a, b] = ['python', 'life']
print(a)
print(b)
```

```python
a = b = 'python'
print(a)
print(b)
```

```python
a = 3
b = 5
a, b = b, a
print(a)
print(b)
```
