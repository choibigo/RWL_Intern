31. 인자에 대해 이터레이션할 때는 방어적이 돼라

```python
def normalize(numbers):
    # (1) 첫 번째 이터레이션 시작
    total = sum(numbers)   
    
    # (2) 두 번째 이터레이션 시작
    result = []
    for value in numbers:
        percent = 100 * value / total
        result.append(percent)
    return result
```

```python
def read_visits_gen():
    yield 15
    yield 35
    yield 80

def normalize_with_print(numbers):
    print("--- normalize 함수 시작 ---")
    
    # 1. 첫 번째 순회: 합계 구하기
    total = sum(numbers)
    print(f"1. 합계(total)를 구했습니다: {total}")
    
    # 2. 두 번째 순회: 비율 구하기
    result = []
    print("2. 각 항목의 비율을 계산하기 위해 루프를 시작합니다.")
    for value in numbers:
        print(f"   현재 값: {value}") # 이터레이터가 소모되었다면 이 줄은 실행되지 않습니다.
        percent = 100 * value / total
        result.append(percent)
    
    print(f"3. 최종 결과 리스트: {result}")
    print("--- normalize 함수 종료 ---\n")
    return result

# 1. 리스트로 넣었을 때 (성공)
print("[테스트 1: 리스트 사용]")
visits_list = [15, 35, 80]
normalize_with_print(visits_list)

# 2. 이터레이터(제너레이터)로 넣었을 때 (실패)
print("[테스트 2: 이터레이터 사용]")
it = read_visits_gen()
normalize_with_print(it)
```

```python
class ReadVisits:
    def __init__(self, data_path):
        self.data_path = data_path

    def __iter__(self):
        with open(self.data_path) as f:
            for line in f:
                yield int(line) # 값을 하나씩 내뱉는 제너레이터를 만듭니다.
```

```python
def my_generator():
    print("제너레이터 시작")
    received = yield 1  # 1을 내보내고, 밖에서 보내준 값을 received에 저장
    print(f"제너레이터가 받은 값: {received}")
    yield 2

it = my_generator()

# 1. 첫 번째 실행 (next를 호출해야 yield까지 갑니다)
first_output = next(it) 
print(f"밖에서 받은 값 1: {first_output}")

# 2. send로 데이터 주입
# 이 값이 yield 식의 결과가 되어 received 변수에 들어갑니다.
try:
    second_output = it.send("안녕!") 
    print(f"밖에서 받은 값 2: {second_output}")
except StopIteration:
    pass
```

```python
def accumulator():
    total = 0
    print("계산기 가동!")
    while True:
        # 1. total을 밖으로 내보내고(출구)
        # 2. 밖에서 send()로 보낸 값을 value에 저장합니다(입구)
        value = yield total 
        
        if value is None:
            break
        
        total += value
        print(f"현재 내부 합계: {total}")

# 제너레이터 객체 생성
calc = accumulator()

# 1. 첫 번째 실행 (반드시 next()나 send(None)으로 첫 yield까지 가야 함)
initial_total = next(calc) 
print(f"처음 받은 값: {initial_total}")

# 2. send로 데이터 '주입'
print(f"결과 1: {calc.send(10)}") # 10을 주입하고, 다음 yield 값인 10을 받음
print(f"결과 2: {calc.send(20)}") # 20을 주입하고, 다음 yield 값인 30을 받음
```

```python
from collections.abc import MutableMapping

# 딕셔너리처럼 작동하지만, 순서를 보장하지 않는 '랜덤 딕셔너리' 클래스
class RandomDict(MutableMapping):
    def __init__(self):
        self._data = {}

    def __setitem__(self, key, value):
        self._data[key] = value

    def __getitem__(self, key):
        return self._data[key]

    def __delitem__(self, key):
        del self._data[key]

    def __iter__(self):
        # ⚠️ 일부러 알파벳 순서대로 정렬해서 내보냅니다 (삽입 순서 무시!)
        return iter(sorted(self._data.keys()))

    def __len__(self):
        return len(self._data)

# 실행해보기
def print_dict(d):
    for key, value in d.items():
        print(f"{key}: {value}")

# 1. 일반 dict (삽입 순서 보장됨)
print("--- 일반 dict 결과 ---")
standard_d = {}
standard_d['banana'] = 1
standard_d['apple'] = 2
standard_d['cherry'] = 3
print_dict(standard_d) 

# 2. RandomDict (삽입 순서 보장 안 됨 - 알파벳순으로 나옴)
print("\n--- RandomDict 결과 ---")
custom_d = RandomDict()
custom_d['banana'] = 1
custom_d['apple'] = 2
custom_d['cherry'] = 3
print_dict(custom_d)
```

```python
def echo_machine():
    print("--- 기계 가동 시작 ---")
    # 1. 처음엔 줄 게 없으니 None을 내보내고 멈춰서 기다림
    # 2. 밖에서 send()로 보낸 값이 received 변수에 저장됨
    received = yield "준비 완료!" 
    
    print(f"기계 내부에서 받은 데이터: {received}")
    yield f"메아리: {received}"

# 1. 제너레이터 객체(기계) 만들기
machine = echo_machine()

# 2. 첫 번째 'next' (기계를 yield 지점까지 이동시키기)
# 이 단계를 거쳐야 yield에서 멈춰서 입력을 기다릴 수 있어요.
status = next(machine)
print(f"밖에서 확인한 상태: {status}")

# 3. 'send'로 데이터 주입하기
# "반가워"라는 데이터를 주입하고, 그 다음 yield가 내놓는 값을 받아옴
result = machine.send("반가워")
print(f"밖에서 확인한 결과: {result}")
```

```python
def adder_generator():
    total = 2
    print("--- 더하기 기계 가동 ---")
    
    while True:
        # 이 한 줄이 입구와 출구 역할을 동시에 합니다!
        # 1. yield total: 현재 합계(total)를 밖으로 내보냅니다.
        # 2. received = yield: 밖에서 send()로 보내준 값을 received에 저장합니다.
        received = yield total
        
        if received is not None:
            total += received
            print(f"전달받은 값: {received}, 현재 합계: {total}")

# 1. 제너레이터 객체 만들기
gen = adder_generator()

# 2. 기계 예열 (첫 번째 yield까지 실행)
# 처음에 total인 0을 내보내고 멈춥니다.
print(f"처음 상태: {next(gen)}") 

# 3. send()로 데이터 주입하기
# 10을 주입하면, 멈춰있던 yield 자리로 10이 들어가서 total에 더해집니다.
print(f"결과 1: {gen.send(10)}")

# 4. 한 번 더 주입하기
print(f"결과 2: {gen.send(20)}")
```

```python
def child():
    # 밖에서 값을 주입받으려고 기다리는 입구
    received = yield
    print(f"  [자식] 주입받은 값: {received}")
    yield f"결과: {received}"

def parent():
    # 자식 두 개를 연결함
    yield from child() # 자식 1호기
    yield from child() # 자식 2호기

# 실행
it = parent()
next(it)  # 첫 번째 자식의 첫 yield까지 예열

print(f"출력 1: {it.send('데이터 A')}")
print("--- 다음 자식으로 넘어가는 지점 ---")
print(f"출력 2: {it.send('데이터 B')}")
```

```python
class MyError(Exception):
    pass

def my_generator():
    yield 1
    try:
        yield 2  # 여기서 밖에서 throw(MyError)를 하면 아래 except로 감
    except MyError:
        print("내부에서 MyError 잡음!")
        yield 3
    yield 4

it = my_generator()
print(next(it))  # 1 출력
print(next(it))  # 2 출력 후 멈춤

# 밖에서 에러를 던짐!
print(it.throw(MyError)) # "내부에서 MyError 잡음!" 출력 후, 다음 yield인 3을 받아옴
```

```python
class SmartCounter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0
        self.error_happened = False  # 예외 상황을 기록하는 '상태 변수'

    def __iter__(self):
        while self.current < self.limit:
            # 상태를 확인해서 동작을 결정함 (상태 전이 로직)
            if self.error_happened:
                yield "에러 발생 수리 모드로 전환."
                self.current = 0     # 에러 시 처음부터 다시 시작하도록 상태 전이
                self.error_happened = False # 수리 완료 후 상태 복구
            else:
                self.current += 1
                yield self.current

# 1. 객체 생성
counter = SmartCounter(limit=5)
it = iter(counter)

# 2. 정상 작동
print(next(it))  # 1
print(next(it))  # 2

# 3. 예외 상황 (throw 대신 속성을 직접 변경)
print("--- 외부에서 문제 감지 ---")
counter.error_happened = True

# 4. 다음 yield 시 기계가 알아서 상태를 보고 동작을 바꿈
print(next(it))  # 에러 발생 수리 모드로 전환.
print(next(it))  # 1 (상태 전이 결과로 다시 1부터 시작)
```
