### 3-3. 다차원 배열의 계산
---

```python
# 2차원 배열
import numpy as np

B = np.array([[1, 2], [3, 4], [5, 6]])
print(B)
print(np.ndim(B)) # 배열의 차원 수를 나타내준다.
print(B.shape)
```




- 행렬의 내적 계산 방법


```python
import numpy as np

A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[1, 2], [3, 4], [5, 6]])
C = np.dot(A, B)
print(B)
```


- 행렬의 곱에서는 대응하는 차원의 원소 수를 일치시켜야 한다.

- 신경망의 내적

```python
import numpy as np

X = np.array([1, 2])
W = np.array([[1, 3, 5], [2, 4, 6]])

Y = np.dot(X, W)
print(Y)
```
