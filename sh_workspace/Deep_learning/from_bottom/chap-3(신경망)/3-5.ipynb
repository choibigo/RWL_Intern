### 3-5. 출력층 설계하기


###### 신경망은 **분류**와 **회귀** 모두에 이용할 수 있는데 목적이 뭔지에 따라 어떤 **활성화 함수**를 쓸 지가 달라짙다.

- 항등 함수: 입력을 그대로 출력한다.
- 소프트맥스 함수: 여러 출력값을 확률로 변환하는 함수다.




```python
# 소프트맥스 함수 구현
import numpy as np

a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)
sum_exp_a = np.sum(exp_a)
y = exp_a / sum_exp_a
print(y)
```

```python
import numpy as np

def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y


# 위와 같이 계산하면 exp_a가 무한히 커지거나 작아져서 overflow가 날수도 있다.
# 이를 방지하기 위해서 아래와 같은 식으로 softmax를 나타낸다.

def softmax2(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y

a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
np.sum(y) # 소프트맥스 함수의 출력은 0에서 1사이의 실수이다. 출력의 총합은 1이다.
```

- 소프트맥스 함수는 단조 증가 함수이기 때문에 입력 원소들 사이의 대소 관계로 그대로 이어진다. 즉, **지수 함수 계산에 드는 자원 낭비를 줄이고자 소프트맥스 함수 사용을 줄이는 경우도 있다.**
