4. Regression, Logistic regression에 대해 설명
- (linear)Regression: 선형 모델로 연속적인 실수형을 예측하는 것. 완벽히 target 값과 같을 수 없기 때문에 그 오차를 통해 가중치,bias를 update (비선형이 필요할 때는 가중치와 bias 연결은 선형으로 하되 입력값을 비선형 함수로 감싼다.)

- Logistic regression: (linear)Regression의 선형 모델 wx + b를 사용하지만, 출력에 Logistic 함수(sigmoid)를 적용하여 결과를 확률로 변환하여 분류 문제를 수행하는 모델.

5. Odds, Logit에 대해 설명



- Odds: 사건이 발생할 확률을 사건이 발생하지 않을 확률로 나눈다. Logistic regression에서는 sigmoid 함수에 Odds 값에 log를 붙힌 Log-Odds를 구해서 넣어줘야 한다.
- 위에서 말한 Log-Odds가 Logit이다. logit은 확률 -> 실수값
그래서 Logistic regression에서 sigmoid에 Logistic 값이 들어간다는 이유가 확률에 대한 실수값이 input으로 들어가기 때문이다.

7. Normalization Layer, Regularization Layer에 대해 설명(Normalization Layer에도 가중치가 있는데 그거에 대해 더 자세히, train loss를 올림으로써 어떻게 overfiting을 막을지)
- Batch Normalization의 한계: batch size에 크게 의존적, RNN, LSTM 같이 sequential한 데이터를 다룰 때는 한계가 있다.(padding 사용)
- Normalization Layer는 batch 단위가 아니라, 각 샘플 단위로 정규화를 수행한다.
- Normalization을 진행한 후에 아래 이미지와 같이 Scailing, Shift를 한다. Scailing, Shift를 추가로 하는 이유는 평균: 0 분산: 1로 만들어 버리면 모델의 표현력을 잃을수도 있기 때문에 감마, 베타를 학습하며 더 좋은 분포를 찾아 나간다.(감마, 베타도 가중치다.) ex) 활성화 함수가 ReLU 같은 경우 정규화만 할 경우 절반이 음수여서 0이되어 사라지기 때문에 Shift를 한다.



- *Nomalization도 하나의 계층으로 보고 감마, 베타에 대한 기울기를 구해서 역전파를 적용시킨다.


- Regularization: train loss를 올림으로써 test loss를 감소시켜서 overfiting을 막는다.

- Regularization layer: Dropout과 같이 Regularization을 하는 신경망 사이의 layer

- Dropout: 역전파 과정은 동일하게 유지 되지만, 예를 들어 h1,h2 뉴런의 출력값이 있다고 가정했을 때 첫 번째 학습에 h1, h2에 랜덤변수(1 or 0)를 곱해서 h1이 비활성화(0이 곱해짐) 되었으면 h2만 살아있는 상태에서 학습되어서 h2에 대한 가중치만 바뀌고, 두 번째는 h2가 비활성화 된다면 h1의 가중치만 학습이 되고 세 번째는 둘 다 활성화 되면 둘 다 학습이 되는 식으로 학습이 진행되며 한 가지 특성이나 뉴런에 의존하지 않고 학습을 할 수 있다.

8. L1,L2 Regularization에 대해 설명




- L1: 기존 loss function에 가중치의 절댓값들(norm)을 더함으로써 가중치가 너무 커지는 것을 방지하고 Loss를 추가함으로써 overfiting되는 것을 막아준다.




- L2: 기존 loss에 L1과 다르게 가중치의 제곱값을 더해줌으로써 L1보다 가중치 값에 더 민감하게 반응한다.

- 둘의 차이: 역전파 관점에서 L1은 미분 했을 때 상수항이 나와서 같은 값으로 계속 밀어 결국 W(가중치)가 0이 될 수 있는데 L2는 미분하면 2W가 나와서 계속해서 W값에 비례해서 작아진다.

- L1, L2를 더함으로써 예를 들어 처음에 Loss 값을 줄이기 위해서 Train값을 맞추기 위해 가중치 값을 높였다고 가정하면 Test 데이터로 Train에 없는 값이 들어왔을 때 가중치가 높으면 그만큼 입력에 민간해질 가능성이 커진다.
기존의 loss function 기울기가 2라고 가정했을 때 기존 가중치 W - (학습률 * 2)만큼 가중치가 줄어들게 된다. 하지만 L2가 1이라고 가정하면 w - (학습률 * 3)으로 가중치가 더 줄어들게 되면서 새로운 Test data가 들어왔을 때 오차가 더 작다.
즉 L1,L2가 가중치가 너무 커지는 것을 억제하는 역할을 해준다.
그러면서 가중치가 너무 커지지 않음으로써 새로운 데이터에 대한 민감도를 낮춰준다.
