### 6-5. 적절한 하이퍼파라미터 값 찾기
###### 하이퍼파라미터를 평가할 때 시험 데이터를 쓰면 하이퍼파라미터가 ***오버피팅*** 될 수 있어서 ***검증 데이터***를 사용해야 한다.
- 검증 데이터를 따로 주어지는 경우도 있지만 없는 경우 훈련 데이터 중 20% 정도를 검증 데이터로 먼저 분리하는 것이다.

---

#### 하이퍼파라미터 최적화

1. 하이퍼파라미터 값의 범위를 설정

2. 설정된 범위에서 하이퍼파라미터의 값을 무작위로 추출

3. 2단계에서 샘플링한 하이퍼파라미터 값을 사용하여 학습하고, 검증 데이터로 정확도를 평가(에폭을 작게 설정)

4. 2,3단계를 반복하며, 그 정확도의 결과를 보고 하이퍼파라미터의 범위를 좁힌다.

### 정리

- 매개변수 갱신 방법에는 확률적 경사 하강법(SGD), 외에도 모멘텀, AdaGrad,Adam 등이 있다.

- 가중치 초깃값을 정하는 방법은 올바를 학습을 하는 데 매우 중요하다.

- 가중치의 초깃값으론는 "Xavier 초깃값"과 'He 초깃값'이 효과적이다.

- 배치 정규화를 이용하면 학습을 빠르게 진행할 수 있으며, 초깃값에 영향을 덜 받게 된다.

- 오버피팅을 억제하는 정규화 기술로는 가중치 감소와 드롭아웃이 있다.
