```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x+y)
print(x-y)
print(x*y)
```

```python
# 넘파이의 N차원 배열
import numpy as np

A = ([[1, 3], [4, 5]])
print(A)
```

```python
# 브로드캐스트: 행렬이 다른 두 배열끼리 계산하여 행렬이 맞춰지는 것.
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])

print(A*B)
```

```python
import numpy as np

X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0])

for row in X:
    print(row)

X = X.flatten()
print(X)
X[np.array([0, 2, 4])]

print(X>15)
print(X[X>15])
```
