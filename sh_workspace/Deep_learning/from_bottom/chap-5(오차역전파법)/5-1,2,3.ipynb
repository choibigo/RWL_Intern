#### 계산 그래프: 계산 과정을 그래프로 나타낸 것.(노드와 에지로 구성)

- ex) 문제: 현빈 군은 슈퍼에서 사과를 2개, 귤을 3개 샀습니다. 사과는 1개에 100원, 귤은 1개 150원입니다. 소비세가 10%일 때 지불 금액을 구하세요.




#### 계산 그래프의 장점

1. 국소적 계산을 통해 아무리 복잡한 계산이라도 간단하게 시행할 수 있다.

2. 계산 그래프는 중간 계산 결과를 모두 보관할 수 있다.

3. *역전파를 통해 **미분**을 효율적으로 계산할 수 있다.

1. 덧셈 노드의 역전파

###### 역전파는 **국소적인 미분**을 순방향과는 반대인 오른쪽에서 왼쪽으로 전달한다.



- 덧셈 노드의 역전파는 입력 값을 그대로 흘려보낸다.

2. 곱셈 노드의 역전파

- 곱셈 노드 역전파는 상류의 값에 순전파 때의 입력 신호들을 '서로 바꾼 값'을 곱해서 보낸다.
