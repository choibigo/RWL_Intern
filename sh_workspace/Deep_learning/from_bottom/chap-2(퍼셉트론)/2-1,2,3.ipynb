### 2-1. 퍼셉트론이란?
###### 퍼셉트론: 다수의 신호를 입력으로 받아 하나의 신호를 출력한다.

- 그림의 원은 **뉴런** 혹은 **노드**라고 부른다.
- 입력 신호가 뉴런에 보내질 때 **가중치**가 곱해지며 **정해진 한계를 넘어설 때만** 1을 출력한다.(이 한계값을 **임계값**이라고 한다.)




---

##### *AND,NAND,OR 게이트가 있고, NAND는 NOT AND를 의미하며 퍼셉트론으로 이런 논리 회로를 표현할 수 있다.

```python
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    if tmp > theta:
        return 1
    
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))

```

```python
import numpy as np

x = np.array([0, 1])
w = np.array([0.5, 0.5])
b = -0.7

print(x*w)
print(np.sum(x*w))
print(np.sum(x*w)+ b)
```

```python
# AND 게이트 구현
import numpy as np

def AND(x1,x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w) + b
    if tmp > 0:
        return 1

    if tmp <= 0:
        return 0

print(AND(1,1))
```
