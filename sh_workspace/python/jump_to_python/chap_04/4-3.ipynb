# 4-3. 파일 읽고 쓰기

```python
f = open("새파일.txt", 'w')
for i in range(1,11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

- 파일이 깨져있는 경우가 있을 수 있다. 이 점을 없애기 위해서는 encoding ="UTF-8"을 넣으면 된다.

```python
f = open("새파일.txt", 'w', encoding = "UTF-8")
for i in range(1,11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
f = open("새파일.txt", 'r')
line = f.readline()
print(line)
f.close()

```

```python
f = open("새파일.txt", 'r')
while True:
    line = f.readline()
    if not line:
        break
    print(line)
f.close()
```

```python
f = open("새파일.txt", 'r')
lines = f.readlines()
print(lines)
f.close()
```

```python
f = open("새파일.txt", 'r')
data = f.read()
print(data)
f.close()
```

### -with 사용
##### with: 자원을 안전하게 사용하고 자동으로 정리해 주는 문법

```python
with open("foo.txt", "w") as f:
    f.write("Life is too short, you need python")
```
