# 4-1. 함수

### 파이썬 함수의 구조

```python
def add(a,b): # a,b는 매개변수
    return a + b

print(add(1,2)) # 3, 4는 인수
```

- 매개변수(parameter) - 함수에서 정의되어 사용되는 변수
- 인수(arguments) - 함수를 호출할 때 건네주는 변수

```python
# 함수의 입력 값이 없을 수도 있다.
def say():
    return 'Hi'

a = say()
print(a)
```

```python
# return 즉 출력이 없는 함수도 있다.

def add(a, b):
    print("%d, %d의 합은 %d입니다."%(a, b, a+b))

add(1,2)

a = add(1,2)
print(a) # return이 없어서 a에 None이 담김
```

```python
def sub(a, b):
    return a - b

a = sub(a=7, b=3)
print(a)
```

```python
# *를 이용해서 우리가 원하는 갯수만큼의 변수를 입력할 수 있다.
def add_many(*args):
    result = 0
    for i in args:
        result = result + i
    return result

add_many(1,2,3,4,5)
```

```python
# **를 이용하면 dictionary로 데이터를 저장하여 key,value로 값을 저장할 수 있다.
def print_kwargs(**kwargs):
    print(kwargs['b'])

print_kwargs(a=1,b=2)
```

- return으로 함수를 빠져나가는 방법은 실제 프로그래밍에서 자주 사용한다.

```python
def example(a):
    if a == '바보':
        return
    print("제대로 실행됐습니다.")

# example('안녕')
example('바보')
    
```

```python
# 매개변수를 미리 정해서 함수를 만들 수도 있다.

def say_myself(name, old, man=True):
    print("나의 이름은 %s 입니다." % name)
    print("나이는 %d살 입니다." % old)
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")

say_myself('한승학', 20)
```

```python
# 함수 안에 있으면 지역변수고 함수 밖에 있으면 전역변수다. 지역변수는 함수 안에서만 사용되고 사라진다.
a = 1

def vartest(a):
    a = a + 1

vartest(3)
print(a)
```

```python
a = 2
def add():
    global a
    a += 3
    return a

print(add())
```

```python
# a는 지역변수여서 사라짐
a = "hello"
def vartest(a):
    a = a + "hi"

vartest(a)
print(a) 
```

```python
# 함수에서 return이 있고 없고의 차이!!

a = [1, 2, 3, 4]
result = a.pop() # pop함수는 return 값이 있다.
print(result)
print(a)

result = a.append(4) # append는 return 값이 없어서 result가 None이 나옴.
print(result)
print(a)

```

- *tip: 함수 위에 마우스를 올려두면 출력 값이 무엇인지 나온다.

- Mutable(가변): 객체를 생성한 후에도 내용을 변경할 수 있다. ex) list, dic, set...
- Immutable(불변) 객체를 생성한 후에는 내용을 변경할 수 없다. ex) int, float, bool...
