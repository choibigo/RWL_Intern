# 2-4. 튜플 자료형
###### 리스트와 튜플의 가장 큰 차이: 리스트는 수정이 가능하고 튜플은 수정이 불가능하다.

- Mutable(리스트, 딕셔너리, 집합): 수정 가능 
- Immutable(정수, 실수, 문자열, 튜플): 수정 불가능

```python
a = [1,2,3]

t1 = (1,) # 튜플은 1개의 요소만을 가질 때는 요소 뒤에 ,를 반드시 붙여야 한다.
t4 = 1, 2, 3, 4 # 이렇게도 튜플 저장 가능하다.
print(type(t1))
print(t1)
print(t4)
```

```python
t1 = (1, 2, 'a', 'b')
print(t1[3])
print(t1[1:])
```

```python
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
t3 = t1 + t2 # 튜플을 변화시킨게 아니라 새로운 튜플을 만든 거기 때문에 가능
print(t3)
```
