# 2-3. 리스트 자료형
###### 리스트(List)는 여러 개의 데이터를 순서대로 저장하는 자료형이다. 즉 데이터를 담는 상자라고 생각하면 된다.

```python
odd = [1, 3, 5, 7, 9]
print(type(odd))
```

```python
e = [1, 2, 'Life', 'is']
print(e[2])
```

```python
# 문자열이어서 더하기가 아니라 연결이 된다.
a = [1, '2', '3']
print(a[1] + a[2])
```

```python
# 리스트 안에 리스트가 들어가 있는 경우가 있다.
a = [1, 2, ['a', 'b', 'c']]
print(a[2][2])
print(a[2])            
```

```python
a = [1, 2, 3, 4, 5]
print(a[::2])
```

```python
a = [1, 2, 3, 4, 5]
b = a[:2]
c = a[2:]
print(b)
print(c)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)
print(a * 3)
print(len(a))

# string 문자열(문자 + 리스트(배열))
```

```python
# 자료형이 다른 것끼리 더하면 에러가 발생한다.
a = [1, 2, 3]
print(str(3) + 'hi')
```

```python
# 리스트의 값 수정하기
a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# del 함수를 사용해서 삭제
a = [1, 2, 3]
del a[1]
print(a)
```

```python
# slicing 기법으로 삭제 가능
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

## 리스트 관련 함수

```python
a = [1, 2, 3]
a.append(4)
a.append([5, 6])
print(a)
```

```python
a = [1, 4, 3, 2]
a.sort()
print(a)

b = ['a', 'c', 'b']
b.sort()
print(b)
```

```python
a = ['a', 'c', 'b']
a.sort()
a.reverse()
print(a)
```

```python
a = [1, 2, 3]
print(a.index(3))
```

```python
a = [1, 2, 3]
a. insert(1, 4) # 인덱스 첫번째 자리에 4를 삽입
print(a)
```

```python
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
print(a)
```

```python
a = [1, 2, 3]
a.pop() # 맨 마지막 요소를 삭제
print(a)
a.pop(1) # 몇 번째 지울지 선택
print(a)
```

```python
a = [1, 2, 3]
a.append([4, 5])
a.extend([6, 7]) # append와 extend차이
print(a)
```
