# 2-3. 리스트 자료형
###### 리스트(List)는 여러 개의 데이터를 순서대로 저장하는 자료형이다. 즉 데이터를 담는 상자라고 생각하면 된다.

```python
odd = [1, 3, 5, 7, 9]
print(type(odd))
```

```python
e = [1, 2, 'Life', 'is']
print(e[2])
```

```python
# 문자열이어서 더하기가 아니라 연결이 된다.
a = [1, '2', '3']
print(a[1] + a[2])
```

```python
# 리스트 안에 리스트가 들어가 있는 경우가 있다.
a = [1, 2, ['a', 'b', 'c']]
print(a[2][2])
print(a[2])            
```

```python
a = [1, 2, 3, 4, 5]
print(a[::2])
```

```python
a = [1, 2, 3, 4, 5]
b = a[:2]
c = a[2:]
print(b)
print(c)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)
print(a * 3)
print(len(a))

# string 문자열(문자 + 리스트(배열))
```
