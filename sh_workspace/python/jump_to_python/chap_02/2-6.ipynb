# 2-6. 집합 자료형

###### 집합(set)은 수학의 집합 개념과 동일한 자료형이다.

```python
s1 = set([1,2,3])
print(s1)
s2 = {'a', 'b', 'c', 'd'}
print(s2)
```

```python
# 집합 자료형에 저장된 값을 인덱싱으로 접급하려면 리스트나 튜플로 변환한 후에 해야 한다.
s1 = set([1,2,3])
l1 = list(s1)
print(l1)

t1 = tuple(s1)
print(t1)
```

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print(s1 & s2) # 교집합 표현
print(s1 | s2) # 합집합 표현
print(s1 - s2) # 차집합 표현

```

```python
s1 = set([1, 2, 3])
s1.add(4)
print(s1) # 하나씩만 추가가 됨
s1.update([4, 4, 5, 6]) # set이기 때문에 중복을 없앰, add와 다르게 한꺼번에 추가할 수 있다.
print(s1)
s1.remove(4)
print(s1)
s1.clear()
print(s1)
```
