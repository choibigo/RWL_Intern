## 2-2. 문자열 자료형

```python
a = "Life is too short, You need Python"
b = "a"
c = "123"
print(type(a))
```

```python
a = "hello world"
b = 'Python is fun'
c = """Life is too short, You need Python"""
d = '''Life is too short, You need Python'''
print(a)
print(b)
print(c)
print(d)
```

```python
# 문자열에 작은따옴표 포함하기
a = "Python's favorite food is perl"
b = 'Python\'s favorite food is perl' # \는 따옴표를 ""안에 있는 걸로 보게 해준다.
print(a)
print(b)
```

```python
a = "Life is \n too short" # \n은 줄바꿈을 의미한다.
print(a)

b = "Life is \t too short" # \t는 tab을 의미한다.
print(b)
```
