# 2-7. 불 자료형

###### 불(bool) 자료형이란 참과 거짓을 나타내는 자료형이다.

```python
a = True # 맨 앞음 대문자로 안 쓰면 오류난다.
print(type(a))

b = 1 > 2
print(b)
```

- [1,2,3] : 참
- () : 거짓
- {'a' : 1} : 참
- {} : 거짓
- 즉 값이 채워져 있으면 참 아니면 거짓

```python
if [1,2,3]:
    print("참")
else:
    print("거짓")
```

```python
a = bool([1,2,3])
print(a)

b = bool([])
print(b)
```
