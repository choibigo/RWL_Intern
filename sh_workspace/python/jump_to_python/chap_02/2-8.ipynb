# 2-8. 자료형의 값을 저장하는 공간, 변수

```python
# 위 그림처럼 a의 주소값을 b에 넣는 거다.
a = [1,2,3]
b = a
print(b)
print(id(b))
print(id(a))

a[1] = 4
print(a)
print(b) # b의 주소값이 똑같기 때문에 a를 바꿔도 b도 같이 바뀐다.
```

```python
# 주소값 복사 말고 값 자체를 복사해서 위와 같은 일이 안 일어나게 할 수 있다.
a = [1, 2, 3]
b = a[:]
a[0] = 4
print(a)
print(b)
```

```python
a,b = ('python' , 'life')
print(a)
print(b)
```

```python
# 파이썬에서는 아래 방법을 이용하여 두 변수의 값을 간단하게 바꿀 수 있다.
a = 3
b = 5
a,b = b,a
print(a)
print(b)
```
