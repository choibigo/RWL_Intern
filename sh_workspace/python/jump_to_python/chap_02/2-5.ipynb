# 2-5. 딕셔너리 자료형
###### 키(key)와 값(Value)를 한 쌍으로 저장하고, 키를 이용해 값을 찾는다. 

```python
dic = {'name' : 'pey' , 'phone' : '010-0000-1234', 'birth' : '1118'}
print(type(dic))
print(dic)
```

```python
a = {1 : 'a'}
a[2] = 'b'
a[3] = [1,2,3] # 리스트도 추가 가능
print(a)
```

```python
a = {1: 'a', 2: 'b', 'name':'pey',3: [1,2,3]}
del a[1]
print(a)
del a['name'] # key 값을 이용해서 value 값을 삭제
print(a)
```

```python
a = {1: 'a', 2: 'b'}
print(a[2]) # key 값을 입력해서 value 값을 알아냄
```

- *key 값이 중복되면 안 되고 key값이 변형되는 값이면(immutable 값이어야 한다.) 안 된다.

## 딕셔너리 관련 함수

```python
# Key 리스트 만들기-keys
a = {'name' : 'pey', 'phone' : '010-9999-1234', 'bitrh' : '1118'}
print(a.keys())
print(a.values())
print(a.items())
```

```python
a = {'name' : 'pey', 'phone' : '010-9999-1234', 'bitrh' : '1118'}
a.clear()
print(a)
```

```python
# get은 찾는 key 값이 없더라도 None이라고 나오고 그 말도 정해줄 수 있다.
a = {'name' : 'pey', 'phone' : '010-9999-1234', 'bitrh' : '1118'}
# print(a['hi']) # 'hi'가 없기 때문에 오류가 발생한다.
print(a.get('hi'))
print(a.get('hi','정보 없음'))
```
