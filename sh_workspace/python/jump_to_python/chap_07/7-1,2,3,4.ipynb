# 클로저와 데코레이터

```python
# 클로저: 함수 안에 함수를 써서 간편하게 함.

def mul(m):
    def wrapper(n):
        return m * n
    return wrapper 

if __name__ == "__main__": # import할 때도 실행되는 걸 막기 위해서 사용
    mul3 = mul(3)
    mul15 = mul(5)

    print(mul3(10))
    print(mul15(10))
```

```python
# 데코레이터: 기존 함수를 바꾸지 않고 기능을 추가할 수 있게 만드는 것.
import time

def elapsed(original_func):   # 기존 함수를 인수로 받는다.
    def wrapper():
        start = time.time()
        result = original_func()    # 기존 함수를 수행한다.
        end = time.time()
        print("함수 수행시간: %f 초" % (end - start))  # 기존 함수의 수행시간을 출력한다.
        return result  # 기존 함수의 수행 결과를 반환한다.
    return wrapper

def myfunc():
    print("함수가 실행됩니다.")

decorated_myfunc = elapsed(myfunc)
decorated_myfunc()
```

```python
import time

def elapsed(original_func):   # 기존 함수를 인수로 받는다.
    def wrapper():
        start = time.time()
        result = original_func()    # 기존 함수를 수행한다.
        end = time.time()
        print("함수 수행시간: %f 초" % (end - start))  # 기존 함수의 수행시간을 출력한다.
        return result  # 기존 함수의 수행 결과를 반환한다.
    return wrapper

@elapsed # 이 방싱으로 데코레이터를 사용할 수도 있다.
def myfunc():
    print("함수가 실행됩니다.")

myfunc()
```

# 이터레이터와 제너레이터

```python
# next()를 호출해야 이터레이턱 다음 값을 하나 꺼내 반환한다.
a = [1, 2, 3]
ia = iter(a) # 이터레이터 생성
print(type(ia))
print(next(ia))
print(next(ia))
print(next(ia))
```

```python
# 제너레이터: 이터레이터를 생성해주는 함수
def mygen():
     yield 'a'
     yield 'b'
     yield 'c'
 
g = mygen()
print(next(g))
print(next(g))
print(next(g))

```

```python
import time

def longtime_job():
    print("job start")
    time.sleep(1)
    return "done"

# 제너레이터 표현식: 함수를 미리 실행하지 않고 필요할 때만 실행
gen_job = (longtime_job() for i in range(5))
print(next(gen_job))  # 첫 번째 값만 요청
```
