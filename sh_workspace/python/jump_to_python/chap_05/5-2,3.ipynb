```python
import [모듈이름] # 이 명령어를 통해 모듈에 있는 클래스나 함수들을 사용할 수 있다.
p
```

```python
import mod1
print(mod1.add(3,4))
print(mod1.sub(4,2))
```

```python
# 모듈 이름을 붙이지 않고 바로 해당 모듈의 함수를 쓸 수 있다. 
from 모듈_이름 import 모듈_함수 
```
