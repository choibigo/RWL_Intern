# 5-5. 내장 함수

###### print() 같이 내장 함수는 너무 많기 때문에 일일이 다 공부하기 보다는 그때 그때 찾아서 사용하는 것이 best!

```python
a = abs(-3) # 절댓값 함수
print(a)
```

```python
all([1,2,3]) # 하나라도 거짓이면 거짓
```

```python
any([0, 0, 0, 0]) # 하나라도 참이면 참 모두가 거짓일 때만 거짓
```

# 5-6. 표준 라이브러리

```python
import datetime

day1 = datetime.date(2021,12,14)
day2 = datetime.date(2023,4,5)

diff = day2 - day1
print(diff.days)
```
