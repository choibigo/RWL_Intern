- 클래스는 과자틀이고 객체는 과자 틀로 찍어 낸 과자다.

```python
class FourCal:
    def setdata(self, first, second): # self를 꼭 써줘야 한다.
        self.first = first
        self.second = second

    def add(self): # self를 꼭 써줘야 한다.
        result = self.first + self.second
        return result

    def multiple(self):
        result = self.first * self.second
        return result

    def divide(self):
        result = self.first / self.second
        return result

a = FourCal()
a.setdata(1,4)
print(a.add())
print(a.multiple())
print(a.divide())

b = FourCal()
b.setdata(3,7)
print(b.add())
print(b.multiple())
print(b.divide())


```

## *class 상속받기

```python
class 클래스_이름(상속할_클래스_이름)

ex) class MoreFourCal(FourCal):
# MoreFourCal 클래스는 FourCal 클래스를 상속했으므로 FourCal 클래스의 모든 기능을 사용할 수 있다.
```

```python
class FourCal: # 부모 class
    
    def __init__(self,first,second):
        self.first = first
        self.second = second

    def setdata(self, first, second): # self를 꼭 써줘야 한다.
        self.first = first
        self.second = second

    def add(self): # self를 꼭 써줘야 한다.
        result = self.first + self.second
        return result

    def multiple(self):
        result = self.first * self.second
        return result

    def divide(self):
        result = self.first / self.second
        return result


class MoreFourCal(FourCal): # 자식 class

    def pow(self):
        result = self.first ** self.second
        return result


a = MoreFourCal(4,2)
print(a.add())
print(a.pow())
```

- overriding: 상속받은 부모의 함수를 자식 클래스에서 재정의하여 사용한다.

```python
class FourCal: # 부모 class
    
    def __init__(self,first,second):
        self.first = first
        self.second = second

    def setdata(self, first, second): # self를 꼭 써줘야 한다.
        self.first = first
        self.second = second

    def add(self): # self를 꼭 써줘야 한다.
        result = self.first + self.second
        return result

    def multiple(self):
        result = self.first * self.second
        return result

    def divide(self):
        result = self.first / self.second
        return result


class MoreFourCal(FourCal): # 자식 class

    def pow(self):
        result = self.first ** self.second
        return result

class SafeFourCal(FourCal):

    def divide(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second

a = SafeFourCal(5,2)
print(a.divide())
```
