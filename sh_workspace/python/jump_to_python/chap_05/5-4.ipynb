# 5-4. 예외 처리

```python

try: # 오류가 발생하는 구문을 입력
    4 / 0
except ZeroDivisionError as e: # 발생하는 오류명 입력
    print("오류 발생") # 오류가 발생했을 때 실행할 명령어
```

```python
try:
    f = open('foot.txt', 'w')
    
finally:
    f.close() # 중간에 오류가 발생하더라도 무조건 실행
```

```python
try:
    age = int(input("나이를 입력하세요: "))

except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print("미성년자는 출입금지입니다.")
    else:
        print("환영합니다.")
```
