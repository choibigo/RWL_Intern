# 3-1. if문

```python
money = True
if money:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```


- if문의 기본 구조(들여쓰기를 잘 해야 됨)



- 조건문에서의 비교 연산자

```python
x = 3
y = 2
print(x != y)
```

```python
money = 2500
if money >= 3000:
    print("택시를 타라")
else:
    print("걸어가라")
```


- and, or, not

```python
money = 1000
card = True
if money >= 2000 or card:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
money = 1000
card = True
if not card:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```


- in, not in 

```python
1 in [1, 2, 3]
```

```python
print('p ' in 'python') # python에서는 공백도 포함한다.
print('p' in 'python')
```

```python
pocket = ['paper', 'cellphone', 'money']
if 'money' not in pocket: # 예약어 pass를 사용함으로써 이 block이 돌아가게 된다면, 그냥 넘어간다.
    pass
else:
    print("걸어가라")
```

```python
# elif 조건문 사용
pocket = ['paper' , 'cellphone']
card = True
if 'money' in pocket:
    print("택시를 타고가라")
elif card:
    print("택시를 타고가라")
else:
    print("걸어가라")
```
