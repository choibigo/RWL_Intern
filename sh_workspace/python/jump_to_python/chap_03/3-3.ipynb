# 3-3. for문

- for문의 기본 구조형

```python
a = [(1,2), (3,5), (5,6)]
for (first , last) in a:
    print(first + last)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number += 1
    if mark >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다."% number)
```

#### reange(start, stop)은 start 이상 stop 미만의 범위를 뜻한다.

```python
for i in range(2,10):
    for j in range(1,10):
        print(i*j, end = " ")
    print(' ')
```
